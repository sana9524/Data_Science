```python
# Converting data to numpy array
X_ary=X.values
X_ary
```

```python
# Normalizing the numerical data
X_norm=scale(X_ary)
X_norm
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

X_pca=pca.fit_transform(X_norm)
X_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1,color='magenta')
```

```python
# Final Dataframe
final_df=pd.concat([data['Type'],pd.DataFrame(X_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(12,8))
sns.scatterplot(data=final_df)
```

```python
from sklearn.preprocessing import normalize

plt.figure(figsize=(18,8))
dendrogram=sch.dendrogram(sch.linkage(X_norm,'complete'))
```

```python
# Create Cluster
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
AgglomerativeClustering(affinity='euclidean', n_clusters=3)
```

```python
y=pd.DataFrame(hc.fit_predict(X_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding cluster to dataset
X_data1=data.copy()
X_data1['clusterid']=hc.labels_
X_data1
```

```python
# within-cluster sum-of-squares criterion
wcss=[]
for i in range(1,6):
  kmeans=KMeans(n_clusters=i,random_state=2)
  kmeans.fit_transform(X_norm)
  wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=0).fit(X_norm)
clusters3
```
